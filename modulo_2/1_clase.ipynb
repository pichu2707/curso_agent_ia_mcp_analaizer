# Creación de un Agente

En estra clase hablaremos de la creación de Agentes SDK de OpenAI. Este SDK de agentes de OpenAI nos ayuda a crear aplkicaciones liggeras y fácil de usar.
Estos agentes cuentan con unos pequeños sets:
* **Agent** son los LLMs equipados con instrucciones y herramientas
* **Handoffs** permitena a los agentes delegar a otros agentes tareas específicas
* **Guardrails** permiten validar las entradas a los agentes
* **Sessions** mantienen automáticamente el historial de conversación entre ejecuciones del agente.
