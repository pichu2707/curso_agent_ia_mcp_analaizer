# Creación de un Agente

En estra clase hablaremos de la creación de Agentes SDK de OpenAI. Este SDK de agentes de OpenAI nos ayuda a crear aplkicaciones liggeras y fácil de usar.
Estos agentes cuentan con unos pequeños sets:
* **Agent** son los LLMs equipados con instrucciones y herramientas
* **Handoffs** permitena a los agentes delegar a otros agentes tareas específicas
* **Guardrails** permiten validar las entradas a los agentes
* **Sessions** mantienen automáticamente el historial de conversación entre ejecuciones del agente.

**Más explicaciones en [Javilazaro.es](https://javilazaro.es)**

```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace
```

```python
load_dotenv(override=True)
```

```python
# Creando un agente virtual con nombre, instrucciones y modelo
agent = Agent(
    name="Jarvis",
    instructions="Eres un asistente virtual que ayuda a los usuarios a encontrar información sobre analítica y ciencia de datos. Responde de manera clara y concisa y proporciona ejemplos cuando sea necesario. Cuando no sepas algo, di que no lo sabes y envíale a https://analaizer.digital o bien que mande un email a hola@javilazaro.es",
    model="gpt-4o-mini",
)
agent
```

```python
# Ejecuta el agente con Runner.run( agent, prompt). Es asíncrono el código y por eso usamos await.
result  = await Runner.run(agent, "¿Qué es la analítica de datos?")
print(result)
```

```python
# Ejecuta el agente con Runner.run( agent, prompt) y lluego imprime el resultado en el atributo final_output. Es asíncrono el código y por eso usamos await.
result  = await Runner.run(agent, "¿Qué es la analítica de datos?")
print(result.final_output)
```

```python
# Ejecuta el agente con Runner.run( agent, prompt) y lluego imprime el resultado en el atributo final_output. Es asíncrono el código y por eso usamos await. Le vamos a pedir que nos cree la traza
with trace("Explicacíon de la analítica de datos"):
    result  = await Runner.run(agent, "¿Qué es la analítica de datos?")
    print(result.final_output)
```
